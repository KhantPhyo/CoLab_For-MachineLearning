```python
#https://www.tutorialspoint.com/matplotlib/matplotlib_quiver_plot.htm
import matplotlib.pyplot as plt
import numpy as np
```

```python
x,y = np.meshgrid(np.arange(-2, 2, .2), np.arange(-2, 2, .25))
x.shape, y.shape
```

```python
z = x*np.exp(-x**2 - y**2)
z.shape
```

```python
#numpy.gradient(f, *varargs, axis=None, edge_order=1)
v, u = np.gradient(z, .2, .2)
v.shape,  u.shape
```

```python
fig, ax = plt.subplots()
q = ax.quiver(x,y,u,v)
plt.show()
```

```python
z = np.sqrt(x**2 + y**2)
z.shape
```

```python
v, u = np.gradient(z)
```

```python
fig, ax = plt.subplots()
q = ax.quiver(x,y,u,v)
plt.show()
```
