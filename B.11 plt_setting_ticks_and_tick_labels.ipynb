```python
#https://www.tutorialspoint.com/matplotlib/matplotlib_setting_ticks_and_tick_labels.htm
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, np.pi*2, 0.05)
```

```python
y = np.sin(x)
```

```python
fig = plt.figure()
#left, bottom, width and height
ax = fig.add_axes([0, 0, 1, 1]) # main axes
ax.plot(x, y)
ax.set_xlabel('angle')
ax.set_title('sine')
plt.show()
```

```python
fig = plt.figure()
#left, bottom, width and height
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
ax.plot(x, y)
ax.set_xlabel('angle')
ax.set_title('sine')
ax.set_xticks([0,2,4,6])
ax.set_xticklabels(['zero','two','four','six'])
ax.set_yticks([-1,0,1])
plt.show()
```
