```python
#https://www.tutorialspoint.com/matplotlib/matplotlib_3d_contour_plot.htm
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, y):
   return np.sin(np.sqrt(x ** 2 + y ** 2))
```

```python
#https://www.w3schools.com/python/python_lambda.asp
f = lambda x, y: np.sin(np.sqrt(x ** 2 + y ** 2))
f
```

```python
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
X.shape, Y.shape, Z.shape
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D contour')
plt.show()
```

```python
n = 1000
theta_max = 8 * np.pi
theta = np.linspace(0, theta_max, n)
x = theta
y =  np.cos(theta)
z =  np.sin(theta)
x.shape, y.shape, z.shape
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot(x, y, z, 'b', lw=2)
ax.plot((-theta_max*0.2, theta_max * 1.2), (0,0), (0,0), color='k', lw=2)
ax.plot(x, y, 0, color='r', lw=1, alpha=0.5)
ax.plot(x, [0]*n, z, color='g', lw=1, alpha=0.5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Helix')
ax.set_axis_off()
plt.show()
```

```python
theta = np.radians(np.linspace(0,360*5,1000))
r = theta**2
x = r*np.cos(theta)
y = r*np.sin(theta)
plt.plot(x,y)
```

```python
t = np.arange(np.pi/2, np.pi, 0.001)
x = np.sin(t)
y = np.cos(t) + np.log(np.tan(t/2))
plt.plot(x,y)
```

```python
#The lemniscate of Bernoulli
t = np.arange(0, 2*np.pi, 0.001)
x = np.divide(np.cos(t),(1+(np.sin(t)) ** 2))
y = np.divide(np.sin(t)*np.cos(t), (1+(np.sin(t)) ** 2))
plt.plot(x,y)
```

```python
#The deltoid curve
t = np.arange(0, 2*np.pi, 0.001)
x = 2*np.cos(t)*(1+np.cos(t))-1
y = 2*np.cos(t)*(1-np.cos(t))
plt.plot(x,y)
```

```python
#The deltoid curve
t = np.arange(0, 2*np.pi, 0.001)
for n in np.arange(1,101):
  x = 2*n*np.cos(t)*(1+np.cos(t))-1
  y = 2*n*np.cos(t)*(1-np.cos(t))
  plt.plot(x,y)
plt.show()
```

```python
#The astroid curve
t = np.arange(0, 2*np.pi, 0.001)
x = (np.cos(t)) ** 3
y = (np.sin(t)) ** 3
plt.plot(x,y)
```

```python
#The astroid curve
t = np.arange(0, 2*np.pi, 0.001)
for n in np.arange(1,11):
  x = n*((np.cos(t)) ** 3)
  y = n*((np.sin(t)) ** 3)
  plt.plot(x,y)
plt.show()
```

```python
#The toroidal spiral
t = np.arange(0, 2*np.pi, 0.001)
x = (4+np.sin(t))*np.cos(t)
y = (4+np.sin(t))*np.sin(t)
z = np.cos(t)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot(x, y, z, 'b', lw=2)
ax.set_axis_off()
plt.show()
```

```python
#The toroidal spiral
t = np.arange(0, 2*np.pi, 0.001)
fig = plt.figure()
ax = plt.axes(projection='3d')
for n in np.arange(1,11):
  x = (4+np.sin(n*t))*np.cos(t)
  y = (4+np.sin(n*t))*np.sin(t)
  z = np.cos(n*t)
  ax.plot(x, y, z, lw=2)
  ax.set_axis_off()
plt.show()
```
